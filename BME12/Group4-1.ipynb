## 1. Load coronavirus file from `dataset/2019_nCoV_data.csv`

```python
from os import path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
import seaborn as sns
color = sns.color_palette()
sns.set()
```

```python
def dp(dataFileName):
    '''
    This function is used to get full data path (dp) from data file name.
    '''
    return path.join('datasets', dataFileName)
```

```python
fileName = '2019_nCoV_data.csv' 
df = pd.read_csv(dp(fileName), encoding='utf-8')
```

## 6. Show dependence between confirmed cases and deaths case

```python
x = df['Deaths']
y = df['Confirmed']
p = np.poly1d(np.polyfit(x,y,1))
p
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
intercept,slope

def linefitline(b):
    return intercept + slope * b
line = linefitline(x)
plt.scatter(x,y)
plt.xlabel('Deaths')
plt.ylabel('Confirmed')
plt.title('dependence between confirmed cases and deaths case')
plt.plot(x,line, c = 'g')
plt.show()
```

## 7. Show dependence between confirmed cases and recovered case

```python
x = df['Confirmed']
y = df['Recovered']
p = np.poly1d(np.polyfit(x,y,1))
p

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
intercept,slope

def linefitline(b):
    return intercept + slope * b
line = linefitline(x)
plt.scatter(x,y)
plt.ylabel('Recovered')
plt.xlabel('Confirmed')
plt.title('dependence between confirmed cases and recovered case')
plt.plot(x,line, c = 'g')
plt.show()
```
